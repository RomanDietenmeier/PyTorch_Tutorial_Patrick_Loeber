#vanishingGradientProblem<br>
Gradient 0 means that neurons are dead and will never be updated!<br>
Thus use leaky ReLU when you notice that your weights do not update!<br>
<br>
Squash output between 0 and 1 which adds up to 1 in total, perfect for probabilities.<br>


```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, 1)
        self.sigmoid = nn.Sigmoid()
        # there is also:
        # nn.sigmoid, nn-softmax, nn.tanh, nn.leaky_relu, nn.elu, nn.prelu, nn.softplus, nn.softsign, nn.hardtanh, nn.hardsigmoid, nn.logsigmoid, nn.softmin, nn.softmax, nn.log_softmax
        # as well as in torch (this list is probably not correct, check the documentation for the most recent list of activation functions):
        # torch.sigmoid, torch.tanh, torch.relu, torch.leaky_relu, torch.elu, torch.prelu, torch.softplus, torch.softsign, torch.hardtanh, torch.hardsigmoid, torch.logsigmoid, torch.softmin, torch.softmax, torch.log_softmax
        # some are in torch.nn.functional:
        # F.sigmoid, F.tanh, F.relu, F.leaky_relu, F.elu, F.prelu, F.softplus, F.softsign, F.hardtanh, F.hardsigmoid, F.logsigmoid, F.softmin, F.softmax, F.log_softmax

    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)
        return out


class AlternativeNeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(AlternativeNeuralNet, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.linear2 = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out = torch.relu(self.linear1(x))
        out = torch.sigmoid(self.linear2(out))
        return out
```
